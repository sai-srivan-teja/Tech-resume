### **EVOLUTION OF MOVIES**

Cinematography is the illusion of movement of the recording and subsequent rapid projection of many still photographic pictures
on a screen. Originally a product of 19th century scientific endeavour, cinema has become a medium of mass entertainment and communication
and today it is a **multi-billion-pound industry**.To be more concrete, no one invented cinema
However in 1891 the Edison Company successfully demonstrated a prototype of the "KINETOSCOPE" which enabled one person at a time to
view moving pictures.The primary question is what were early films like? The answer to this question is At first,films were very short sometimes only a few minutes or less. They were shown at fairgrounds,music halls, or anywhere a screen could be set up and a room darkened. Subjects included local scenes and activities, views of foreign lands, short comedies and newsworthy events. The first 30 years of cinema were characterised by the growth and consolidation of an industrial base, the establishment of the narrative form and refinement of technology. 1911-1926 THE SILENT ERA 

1927-1940 THE SOUND ERA

# **MODERN FLIM INDUSTRIES AND FLIMMAKING**

After having a quick information about the evolution of movies now lets talk about our modern film industry.
The modern film industry comprises the technology and commercial institutions of filmmaking,i.e., film production companies, film studios, 
cinematography, animation, film production,screenwriting, pre-production,post production, film festivals, distribution and actors.
 
 ### FILMMAKING:

1.CONCEPT AND IDEA GENERATION ":   A film concept is the very basic thought of what it can be. Its the spark of a story or situation and the central characters you will follow.

2.BUDGETING: To spend money on your film carefully, you will make a budget. Decide a realistic limit on what you can afford early enough and track spending throughout the film making process.  

3.IDEA DEVELOPMENT: The development of film ideas is a repeatable step. It should happen throughout the film production process to refine the quality of your production along the way.

4.SCREENWRITING AND SCRIPT MAKING: Script writing is mostly about dialog and how the various characters speak. Descriptions of place or situation set the scenes but primarily you want a script that tells the stroy through their words.

5.HIRING,RECRUITING CAST AND CREW: You want to find the best people for each role that the budget affords.Also Hold script read-throughs.
Audition and screen-test actors.Watch showreels to check technical skills of new crew members.

6.SCOUTING LOCATIONS AND PRODUCTION DESIGNS: Finding suitable locations for the film shoot is known as scouting. Scouts search for interior and outdoor places that best fit those described in the script.Get permission to film. Shoot in one place or use  green screening

7.STORYBOARDS AND SHOT LISTS: Storyboards are like a script for the camera to follow.Instead of words they describe key scene shots and camera anglers with illustrations.

8.PRODUCTION SCHEDULES: A production schedule helps with keeping the film production process organized. It typically takes the form of a day-by-day calendar, spreadsheet, or chart that timetables the filming. With the project all planned production can begin.This requires filming all the scenes you've scheduled , including both audio and visuals

9.SHOOTING: Shooting or filming is the main single step here. Its all about capturing the raw video all the shots and sounds that the script demands.

10.EDITING: The film editor's job is to join together the various shots into scenes, and the scenes into a finished movie.Thefilm editor must creatively work with the layers of images, story, dialogue, music,pacing.

11.SOUND DESIGN: This step is typically about the creation of additional sounds not recorded on set. Sound design typically includes adding extra sound effects including 'diegetic sounds' (sounds that exist in the scene) or non-diegetic sounds (sounds that area used for emphasis or to aid in telling the story).

12.COLOR CORRECTION: Color correction is the process of adjusting the colours in the footage to make it look more "true-to-true'. When colour correcting footage we just want to make sure that everything is exposed correctly,there are no weird colour casts and that skin tones and other colours in the scene look accurate.

13.VISUAL EFFECTS: Visual effects or 'VFX' is the adding of images to the film that were impossible or impractical to shoot in the production stage.VFX artists use software to add or remove objects, people, places, and fireballs.

14SOUND MIXING: Sound mixing ensures every audio elements within a film's soundtrack is correctly balanced.The primary goal is to set the volume levels to achieve clarity.

15.COLOUR GRADING: Colour grading is more artistic and is used to style a film's picture colour in less natural and more intentional ways.Every frame can be re-tined digitally using certain palettes or filters.

16.DISTRIBUTION: Finished films are marketed and shown by a film distributor.Separate from the film's main production team, the distributor releases the film to the public.
**DISNEY, WARNER BROS, SONY AND UNIVERSAL** are top film distributors in the trade

17.PROMOTION Promoting a film is any work done to make people aware of it. This includes the marketing campaigns that are usually devised and run by the film's distributor. However most film promotion starts with a preview of a trailer.

## 9) Number of Films of Each Genre 

```python
genre_count = data.groupby('Main_Genre').title.count().sort_values(ascending=False)

plt.figure(figsize=(14,8), dpi=120)

with sns.axes_style('darkgrid'):
    fig = sns.barplot(x=genre_count.index, 
                y=genre_count.values)
    fig.set(xlabel="Main Genre", ylabel="Count", title="Number of Films of Each Genre")
    plt.show(fig)
```

**Most number of movies fall under the thriller genre**

## 10) Rating Count

```python
data_copy = pd.DataFrame(data)
data_copy['rating_count'] = data_copy['rating']
genre_rating = data_copy.groupby(by=['Main_Genre','rating'], as_index=False).agg({'rating_count': pd.Series.count})
 
genre_rating = genre_rating.sort_values(ascending=False, by='rating_count')

genre_rating.head()
```

```python
fig = px.bar(genre_rating, x="rating", y="rating_count", color="Main_Genre", barmode = 'stack')
 
fig.show()
```

**PG-13 flims have highest rating count and G rated movies have lowest rating count**

## 11) Mean Length v/s Genre

```python
genre_length = data.groupby('Main_Genre').length.mean().sort_values(ascending=False)

plt.figure(figsize=(14,8), dpi=120)

plt.title("Mean Length v/s Genre", fontsize=14)
plt.xlabel("Main Genre", fontsize=14)
plt.ylabel("Length of Film")

plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.grid(color="grey", linestyle="--", alpha=0.3)
plt.bar(genre_length.index, genre_length.values)
```

**War and Action movies are lengthy compared to comedy and animation**
